☼ Write code that removes whitespace at the beginning and end of a string, and normalizes whitespace between words to be a single space character.  
do this task using split() and join()

```python
def normalize(s: str):
    s = " ".join(s.rstrip().lstrip().split())
    return s
```

```python
normalize("  love    and  hate ")
```

do this task using regular expression substitutions

```python
text = " love    and  hate   "
nltk.re.sub(r"(^\s+)|\s(?=\s)|(\s+$)", "", text) #space at beginning of str OR pos lookahead 2 spaces OR end of str
```

◑ Write code to initialize a two-dimensional array of sets called word_vowels and process a list of words, adding each word to word_vowels[l][v] where l is the length of the word and v is the number of vowels it contains.

```python
words = nltk.corpus.brown.words(categories = "humor")

longest_w = sorted(set([w for w in words]), key = lambda x: len(x))[-1]
n, m = len(longest_w), len(re.findall(r"[AEIOUaeiou]", longest_w))

word_vowels = [["" for vow in range(m+1)] for length in range(n+1)] #n29, m11

for w in words:
    length = len(w)
    vows = len(re.findall(r"[AEIOUaeiou]", w))
    word_vowels[length][vows] = w

print(word_vowels)    
```

◑ Write a function novel10(text) that prints any word that appeared in the last 10% of a text that had not been encountered earlier.

```python
def novel10(text: list):
    cnt = 0
    nineth = int((len(text)*.9))
    rest, tenth = text[:nineth], text[nineth:]
    return set([w for w in tenth if w not in rest])
```

```python
raw = nltk.corpus.gutenberg.raw('melville-moby_dick.txt')
tokens = nltk.word_tokenize(raw)
novel10(tokens)
```
