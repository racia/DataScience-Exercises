```python
### Bigrams use
files = movie_reviews.fileids()
pos, neg = [], []
# Get sentences
for p_review, n_review in zip(files[-7:], files[:7]):
    pos.append(movie_reviews.sents(f"{p_review}"))
    neg.append(movie_reviews.sents(f"{n_review}"))
# convert to 2d list    
pos = [e for sl in pos for e in sl]
neg = [e for sl in neg for e in sl]
```

```python
# Get bigrams for sentences

bigrams_p, bigrams_n = [], [] # [[(" ", " "), ("", ""), ...], ...]

for p, n in zip(pos,neg):
    p_s = list(nltk.bigrams(p))
    n_s = list(nltk.bigrams(n))
    p_s[:] = [b for b in p_s]# if " ".join(b).replace(" ", "").isalpha() and b not in stopwords.words("english")]
    bigrams_p.append(p_s)
    n_s[:] = [bb for bb in n_s]# if " ".join(bb).replace(" ", "").isalpha() and bb not in stopwords.words("english")]
    bigrams_n.append(n_s)
    

# format to lowercase
for l, ll in zip(bigrams_p, bigrams_n):
    for t, tt in zip(l, ll):
        str1, str2 = " ".join(t), " ".join(tt)
        l.remove(t)
        ll.remove(tt)
        t = tuple((str1.lower().split()))
        tt = tuple((str2.lower().split()))
        l.append(t)
        ll.append(tt)

# Filter shared bigrams and save sentence indices each
shared = []
p_index, n_index = {}, {}
for p in bigrams_p:
    for t in p:
        for n in bigrams_n:
            if t in n and t not in shared:
                n_index[t] = bigrams_n.index(n)
                p_index[t] = bigrams_p.index(p)
                shared.append(t)
# same length
#assert len(shared) == len(p_index) == len(n_index)
```

```python
# Compare sentences of bigrams

def get_context(bigrams, index):
    return bigrams[index]

for b in shared[:1]:
    print(f"Positive: {get_context(bigrams_p, p_index[b])}")
    print(f"Negative: {get_context(bigrams_n, n_index[b])}\n")
```

```python
from itertools import chain
from nltk.corpus import stopwords
import random

bgr_p = [e for sl in bigrams_p for e in sl]
bgr_n = [e for sl in bigrams_n for e in sl]

cfd_p = CFD(bgr_p)
cfd_n = CFD(bgr_n)

# Compare context of bigrams
def generate_model(cfdist, word, num=4):
    l = []
    while len(l) < num:
        word = random.choice(list(cfdist[word]))
        if word not in l and word not in stopwords.words("english"):
            l.append(word)
    return l

i = 0
for b in shared:
    if b[i] not in stopwords.words("english") and b[i].isalpha():
        print(f"Positive context for '{b[i]}':\n {generate_model(cfd_p, b[i])}")
        print(f"Negative context for '{b[i]}':\n {generate_model(cfd_n, b[i])} \n")
        i = 1 - i
```

◑ Define a function supergloss(s) that takes a synset s as its argument and returns a string consisting of the concatenation of the definition of s, and the definitions of all the hypernyms and hyponyms of s.

```python
def supergloss(s):
    defn = f"Definition: {s.definition()}"
    for hpe in list(s.hypernyms()):
        defn += f" Hypernym: {hpe.name()}: {hpe.definition()}"
    for hpo in list(s.hyponyms()):
        defn += f" Hyponym: {hpo.name()}: {hpo.definition()}"
    return defn
```

◑ Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
import itertools

def top_50(text):
    fdist = FreqDist(w.lower() for w in text if w not in stopwords.words("english"))
    fdist = dict(sorted(fdist.items(), key= lambda x: x[1], reverse=True))
    return dict(itertools.islice(fdist.items(), 50))
    
```

◑ Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
def top_bigrams(text):
    bigrams = list(nltk.bigrams(text))
    d = dict.fromkeys([b for b in bigrams if b[0].lower() not in stopwords.words("english") and b[1].lower() not in stopwords.words("english")])
    for b in d:
        d[b] = bigrams.count(b)
    top_bigrams = dict(sorted(d.items(), key= lambda x: x[1], reverse=True))
    return dict(itertools.islice(top_bigrams.items(), 50))
```

◑ Define a function hedge(text) which processes a text and produces a new version with the word 'like' between every third word.

```python
def hedge(text: str):
    like = []
    for i, w in enumerate(text):
        if i%3 == 0 and i != 0:
            like.append("like")
            like.append(w)
        else:
            like.append(w)
    return like
```

★ What is the branching factor of the noun hypernym hierarchy? I.e. for every noun synset that has hyponyms — or children in the hypernym hierarchy — how many do they have on average? You can get all noun synsets using wn.all_synsets('n').

```python
def branch_factor() -> int:
    cnt = 0
    syn = list(s for s in wn.all_synsets("n") if s.hyponyms())
    for s in syn:
        cnt += len(s.hyponyms())
    avg = round(cnt / len(syn), 1)
    return avg
            
```

★ The polysemy of a word is the number of senses it has. Using WordNet, we can determine that the noun dog has 7 senses with: len(wn.synsets('dog', 'n')). Compute the average polysemy of nouns, verbs, adjectives and adverbs according to WordNet.

```python
def avg_polysemy() -> int:
    POS = ["n", "a", "v"] #Noun -Adjective/Adverb - Verb
    for p in POS:
        cnt = 0
        for w in list(wn.all_lemma_names(f"{p}")):
            cnt += len(wn.synsets(w, f"{p}"))
        avg = round(cnt/len(list(wn.all_lemma_names(f"{p}"))), 2) 
        print(f"Average polysemy count in {p}: {avg}")
```

```python
import numpy as np
from matplotlib.pylab import Figure
words = ["car-automobile", "gem-jewel", "journey-voyage", "boy-lad", "coast-shore", "asylum-madhouse", "magician-wizard", "midday-noon", "furnace-stove", "food-fruit", "bird-cock", "bird-crane", "tool-implement", "brother-monk", "lad-brother", "crane-implement", "journey-car", "monk-oracle", "cemetery-woodland", "food-rooster", "coast-hill", "forest-graveyard", "shore-woodland", "monk-slave", "coast-forest", "lad-wizard", "chord-smile", "glass-magician", "rooster-voyage", "noon-string"]

g = (w.split("-") for w in words)
#print("Path Similarity between: ")
#  for plotting:
x, y = [],[]
for i in range(len(words)):
    pair = next(iter(g))
    a = globals()[pair[0]] = wn.synsets(pair[0])[0]
    b = globals()[pair[1]] = wn.synsets(pair[1])[0]
    sim = round(a.path_similarity(b), 3)
    #(f"{pair[0]} and {pair[1]}: {sim}")
    x.append(str(pair))
    y.append(sim)

#plot:    
fig, ax = plt.subplots()
fig.set_figwidth(20)
fig.set_figheight(10)
plt.xticks(np.arange(len(words)), words, rotation = 25)

# Miller & Charles, 1998 ranking:
plt.plot(x, y, label = "Miller & Charles, 1998 ranking")

# Wordnet ranking:
yy = y.copy()
y.sort(reverse=True)
ax.scatter(x, y)
for i, txt in enumerate(x):
    ax.annotate(txt,(x[yy.index(y[i])], y[i]), rotation=10)
plt.plot(x, y, label = "WordNet Ranking")
plt.legend()
```
